# Functions

```python
G2 = 0.05
G1 = 0.1
def H(x):
    return 0.5*(1+G2/G1)+0.5*(1-G2/G1)*j0(2*x)
def ϕ(k,x):
    return 0.5/H(x)*j0(x)*jn(k,x)
def I(k,xs):
    return array([ϕ(k,x)**2+ϕ(-k,x)**2 for x in xs])
```

```python
xs = linspace(0,10,100001)
```

# Plotting
