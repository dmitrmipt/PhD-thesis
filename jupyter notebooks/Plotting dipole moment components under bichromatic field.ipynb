# Preamble

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from numpy import *
from scipy.special import *
```

```python
from sympy import lambdify
```

# Functions

```python
G1 = 0.1
G2 = G1/2
def H(x):
    return 0.5*(1+G2/G1)+0.5*(1-G2/G1)*j0(2*x)
def ϕ(k,x):
    return 0.5/H(x)*j0(x)*jn(k,x)*G1/k
def I(k,xs):
    return array([ϕ(k,x)**2+ϕ(-k,x)**2 for x in xs])
```

```python
xs = linspace(0,20,10001)
```

# Plotting 

```python
ks = [1,3,5,7,9,11]
plt.figure(figsize=(6,3))
for k in ks:
    plt.semilogy(xs, I(k,xs))
plt.legend(ks)
plt.ylim((1e-8,2e-1))
plt.xlabel(r'$2\Omega/\delta\omega$')
plt.tight_layout()
plt.savefig('zeros.pdf', format='pdf', dpi=400)
```
